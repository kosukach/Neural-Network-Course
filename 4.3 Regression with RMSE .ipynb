```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.callbacks import EarlyStopping
from sklearn import metrics


df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/jh-simple-dataset.csv",
    na_values=['NA','?'])

df.drop("id", axis=1, inplace=True)
x_cols = df.columns.drop(["job", "area", "product", "income"])
for col in x_cols:
    df[col] = zscore(df[col])
df.head()
```

```python
dummies = pd.get_dummies(df["area"], prefix="area")
df = pd.concat([df, dummies], axis=1)
df.drop("area", axis=1, inplace=True)

dummies = pd.get_dummies(df["job"], prefix="job")
df = pd.concat([df, dummies], axis=1)
df.drop("job", axis=1, inplace=True)

dummies = pd.get_dummies(df["product"], prefix="product")
df = pd.concat([df, dummies], axis=1)
df.drop("product", axis=1, inplace=True)
```

```python
nonna = df[df["income"].notna()]
isna = df[df["income"].isnull()]
nx = nonna.drop(["income", "age"], axis=1).values
ny = nonna["income"].values
nx_train, nx_test, ny_train, ny_test = train_test_split(nx, ny, test_size=0.2)
reg = LinearRegression().fit(nx_train, ny_train)
isna["income"] = reg.predict(isna.drop(["income", "age"], axis=1).values)
df = pd.concat([isna, nonna], axis=0)
reg.score(nx_test, ny_test)
```

```python
df["income"] = zscore(df["income"])
```

```python
x = df.drop("age", axis=1)
y = df["age"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = Sequential()
model.add(Dense(25, input_dim=x.shape[1], activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(1))
model.compile(loss="mean_squared_error", optimizer="adam")
monitor = EarlyStopping(monitor="val_loss", min_delta=1e-3, patience=5,
                        verbose=1, mode="auto", restore_best_weights=True)
model.fit(x, y, validation_data=(x_test, y_test), callbacks=[monitor], verbose=2, epochs=1000)
```

```python
pred = model.predict(x_test)

score = metrics.mean_squared_error(y_test, pred)
print("MSE: {}".format(score))
score = score**0.5
print("RMSE: {}".format(score))
```

```python
def lift_chart(y, pred, sort=True):
    t = pd.DataFrame({"pred": pred, "y": y.flatten()})
    if sort:
        t.sort_values(by=["y"], inplace=True)
    plt.plot(t["y"].tolist(), label="expected")
    plt.plot(t["pred"].tolist(), label="predicted")
    plt.ylabel("output")
    plt.legend()
    plt.show()
```
