```python
taxdf = pd.read_csv('https://www.irs.gov/pub/irs-soi/16zpallagi.csv')
```

```python
taxdf = taxdf.loc[(taxdf['zipcode'] != 0) & (taxdf['zipcode'] != 99999),['STATE', 'zipcode', 'agi_stub', 'N1']]
```

```python
taxdf[0:5]
```

```python
medians = {1:12500,2:37500,3:62500,4:87500,5:112500,6:212500}
taxdf['agi_stub'] = taxdf.agi_stub.map(medians)
```

```python
taxdf[0:5]
```

```python
groups = taxdf.groupby(by='zipcode')
```

```python
agidf = pd.DataFrame(groups.apply(lambda x:sum(x['N1']*x['agi_stub'])/sum(x['N1']))).reset_index()
```
