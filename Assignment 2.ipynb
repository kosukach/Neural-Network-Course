```python
import os
import pandas as pd
from scipy.stats import zscore


df = pd.read_csv("http://data.heatonresearch.com/data/t81-558/datasets/reg-33-data.csv")

df = df[0:1000]
df
```

```python
df["ratio"] = df.apply(lambda x: x["max"]/x["number"], axis=1)
```

```python
df.head()
```

```python
cat_dum = pd.get_dummies(df["cat2"], prefix="cat2_")
```

```python
df = pd.concat([df, cat_dum], axis=1)
```

```python
df.drop(["cat2"], axis=1, inplace=True)
```

```python
df.head()
```

```python
item_dum = pd.get_dummies(df["item"], prefix="item_")
```

```python
df = pd.concat([df, item_dum], axis=1)
```

```python
df["length"].median()
```

```python
df["length"].fillna(df["length"].median(), inplace=True)
```

```python
df["height"].fillna(df["height"].median(), inplace=True)
```

```python
df["height"] = zscore(df["height"])
df.columns
```

```python
df.drop(["id", "convention", "usage", "region", "code", "power", "weight", "country", "target"], axis=1, inplace=True)
```
